```python
from sklearn.model_selection import train_test_split

X = datasets.data
#X = df['RM']

y = datasets.target #주택 가격


X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.33, random_state=42)

print(X.shape, X_train.shape, X_test.shape)
```

```python
import numpy as np

X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
#reshpe(-1,1) 더알아보기..
y_train = np.array(y_train).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
```

```python
y_train.shape
```

```python
import numpy as np

X_train = np.array(X_train)
X_test = np.array(X_test)

y_train = np.array(y_train).reshape(-1, 1)
y_test = np.array(y_test).reshape(-1, 1)
```

```python
y_train.shape
```

## 1-4. 정규화 하기

### Z score Normalization:  $\frac{X - \mu}{\sigma}$ 

### min-max Normalization: $\frac{X - X_{min}}{X_{max} - X_{max}}$

```python
from sklearn.preprocessing import StandardScaler

standardScaler = StandardScaler()

standardScaler.fit(X_train) # mean, std

# 모델을 학습하고 계속 사용을 해야돼 
X_scaling_train = standardScaler.transform(X_train)
X_scaling_test = standardScaler.transform(X_test)
```

```python
X_scaling_train
```

## 1-5. tensor로 매핑하기

```python
import torch

X_train_tensor = torch.tensor(X_scaling_train, dtype=torch.float)
y_train_tensor = torch.tensor(y_train, dtype=torch.float).view(-1, 1) #view == reshape

X_test_tensor = torch.tensor(X_scaling_test, dtype=torch.float)
y_test_tensor = torch.tensor(y_test, dtype=torch.float).view(-1, 1)
```

# 2. Model 정의 (Linear Regression)

$y = Wx + b$

```python
import torch
import torch.nn as nn

class LinearRegression(nn.Module):
    def __init__(self, input_size, output_size):
        super(LinearRegression, self).__init__()  # 이단계 까지 어떤 모델을 쓰던 모두 동일하다. 
        self.linear1 = nn.Linear(input_size, output_size)
    
    def forward(self, X):
        y = self.linear1(X)
       
        return y

model = LinearRegression(13, 1)
```

```python
model
```

# 3. Loss Function, Optimizer 정의

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

# 4. 학습 하기

```python
epochs = 1000

model.train()  # 평가하고 싶을 떈, model.eval()
# loss_train = []  # 잘 쌓이고있는지 확인하기 위함 

for epoch in range(epochs):
    pred = model(X_train_tensor)
    
    loss = criterion(pred, y_train_tensor)
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    print(loss.item())
```

```python
#epochs 다시 지정하려면 model을 초기화 하고 재진행 해야함 
epochs = 10000

model.train()
loss_train = []

for epoch in range(epochs):
    prediction = model(X_train_tensor)
    
    loss = criterion(prediction, y_train_tensor)
    
    optimizer.zero_grad()
    loss.backward()
    
    optimizer.step() #gradient descent
    
    
    print(f"Epoch[{epoch}/{epochs}] loss:{loss.item():.3f}")
```
